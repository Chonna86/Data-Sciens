```python
missing_values = df.isnull().sum() / df.shape[0]
print(missing_values)
```

```python
columns_to_check = df.columns.difference(['Мова програмування'])
df_cleaned = df.dropna(subset=columns_to_check)
print(df_cleaned.head())
```

```python
cleaned_shape = df_cleaned.shape


missing_values_percentage = df_cleaned.isnull().mean() * 100

print("Розмір оновленого датафрейму:", cleaned_shape)
print("Частка пропусків у кожному стовпці:")
print(missing_values_percentage)
```
