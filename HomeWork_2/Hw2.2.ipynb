```python
missing_values = df.isnull().sum()/df.shape[0]
print(missing_values)
```

```python
columns_to_check = df.columns.difference(['Мова програмування'])
df_cleaned = df.dropna(subset=columns_to_check)
print(df_cleaned.head())
```

```python
cleaned_shape = df_cleaned.shape


missing_values_percentage = df_cleaned.isnull().mean() * 100

print("Розмір оновленого датафрейму:", cleaned_shape)
print("Частка пропусків у кожному стовпці:")
print(missing_values_percentage)
```

```python
grouped_by_position = df.groupby('Посада')

for position, group in grouped_by_position:
    print("Посада:", position)
    print(group.head())
    print()


fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

for position, group in grouped_by_position:
    x = group.index
    y = group['Зарплата на місяць']
    z = group['exp']
    ax.plot_trisurf(x, y, z, linewidth=0.2)

ax.set_xlabel('Індекс рядка')
ax.set_ylabel('Зарплата на місяць')
ax.set_zlabel('Досвід роботи')
ax.set_title('Діаграма з конусами')

plt.show()
```

```python
file_path = "D:/PythonProjects/MY_PYTHON_WAY/PYTHON_DATA-SCIENS/HomeWork_2/2017_jun_final - 2017_jun_final.csv"
df = pd.read_csv(file_path)
aggregated_salary = df.groupby('Посада')['Зарплата на місяць'].agg(['min', 'max'])
print(aggregated_salary)

plt.figure(figsize=(10, 8))
sns.set(style="whitegrid")
sns.barplot(x=aggregated_salary.index, y='min', data=aggregated_salary, color='skyblue')
sns.barplot(x=aggregated_salary.index, y='max', data=aggregated_salary, color='salmon')
plt.title('Мінімальна та максимальна зарплата за посадами')
plt.xlabel('Посада')
plt.ylabel('Зарплата на місяць')
plt.xticks(rotation=90)
plt.legend(['Мінімальна зарплата', 'Максимальна зарплата'])
plt.show()
```

```python
def fill_avg_salary(row):
    return df['Зарплата на місяць'].mean()

df['avg'] = df.apply(fill_avg_salary, axis=1)
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['avg'], color='blue', alpha=0.5)
plt.title('Середня зарплата для кожного рядка')
plt.xlabel('Індекс рядка')
plt.ylabel('Середня зарплата')
plt.grid(True)
plt.show()
```
